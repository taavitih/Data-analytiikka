```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use("fivethirtyeight")
```

Tarkastellaan kahta "lempiosakettani" Helsingin pörssistä, Kone ja Sampo. Molempien osakkeiden hinnat ovat sattumoisin pyörineet viimeisen viiden vuoden aikana, suurinpiirtein samassa koko luokassa, joten niiden vertailu on helpohkoa, vaikka eivät samalla toimialalla olekaan.

```python
kone = pd.read_csv("KNEBV-2014-02-27-2019-02-27.csv", sep=";", decimal=",", skiprows=2, names=["Date", "Bid", "Ask", "Opening price", "High price", "Low price", "Closing price", "Average price", "Total volume", "Turnover", "Trades", "X"])
kone.head(10)
```

```python
kone=kone.drop(columns="X")
kone.index=pd.to_datetime(kone["Date"], dayfirst=True)
kone=kone.drop(columns="Date")
kone.head()
```

```python
sampo = pd.read_csv("SAMPO-2014-02-27-2019-02-27.csv", sep=";", decimal=",", skiprows=2, names=["Date", "Bid", "Ask", "Opening price", "High price", "Low price", "Closing price", "Average price", "Total volume", "Turnover", "Trades", "X"])
sampo.head(10)
```

```python
sampo=sampo.drop(columns="X")
sampo.index=pd.to_datetime(sampo["Date"], dayfirst=True)
sampo=sampo.drop(columns="Date")
sampo.head()
```

```python
kone["Closing price"].plot(figsize=(15,6))
sampo["Closing price"].plot(figsize=(15,6))
plt.legend(["Kone", "Sampo"])
```

Tässä vertailtuna Koneen ja Sammon osakkeiden kehitystä viimeisen viiden vuoden ajalta.

```python
kone["Closing price"].plot(figsize=(15,6), color="C0")
kone["Closing price"].rolling(100).mean().plot(color="C3")
sampo["Closing price"].plot(figsize=(15,6), color="C1")
sampo["Closing price"].rolling(100).mean().plot(color="C5")
plt.legend(["Kone", "Kone ka", "Sampo", "Sampo ka"])
```

Tarkasteltaessa 100 päivän juoksevaa keskiarvoa ovat molemmat osakkeet olleet vuoden 2018 keskivaiheilla yli "keskiarvoviivan", mikä tarkoittaa että osake on niin sanotusti kallis. Vuoden vaihteessa 2018-2019 osakkeiden hinnat laskivat reilusti ja täten painuivat todennäkäisesti "keskiarvoviivan" alapuolelle, hinnaltaa "edullisiksi". Vuonna 2019 osakkeiden hinnat ovat kuitenkin nousseet taas, joten todennäköisesti osakkeet ovat muuttuneet taas "kalliiksi". Tähän saa kuitenkin täyden varmuuden vasta kun ollaan menty tästä hetkestä 100 päivää eteenpäin.

```python
kone["Trades"][:"2018-01"].plot(figsize=(15,6))
sampo["Trades"][:"2018-01"].plot(figsize=(15,6))
plt.legend(["Kone", "Sampo"])
```

Kun tarkastellaan osakkeilla tehtyjen kauppojen määrää, niin huomataan hetkellisiä piikkejä molemmissa, mutta volyymit silti samaa luokkaa.

Ja seuraavaksi voisin tarkastella vielä mielenkiintoni herättäneen yhtiön, Huhtamäen osakkeen kehitystä viimeisen viiden vuoden ajalta.

```python
huhta = pd.read_csv("HUH1V-2014-02-27-2019-02-27.csv", sep=";", decimal=",", skiprows=2, names=["Date", "Bid", "Ask", "Opening price", "High price", "Low price", "Closing price", "Average price", "Total volume", "Turnover", "Trades", "X"])
huhta=huhta.drop(columns="X")
huhta.index=pd.to_datetime(huhta["Date"], dayfirst=True)
huhta=huhta.drop(columns="Date")
huhta.head(10)
```

```python
huhta["Closing price"].plot(figsize=(15,6), color="C1")
huhta["Closing price"].rolling(100).mean().plot(color="C2")
huhta["Closing price"].rolling(200).mean().plot(color="C3")
plt.legend(["Huhtamäki", "ka 100", "ka 200"])
```

Pikaisen analyysin pohjalta, tässä voidaan tulkita, että lähes koko vuoden 2018 ajan Huhtamäen osake oli juoksevan keskiarvon mukaan "kallis", katsoi sitten 100 tai 200 päivän keskiarvoa. Loppuvuodesta 2018 tuli kuitenkin syvempi niiaus, jolloin todennäköisesti osakkeen hinta vaihtui tämän arvion mukaan "edullinen"-puolelle. Vuoden 2019 alun kehitys mahdollisesti heilautti osakkeen jälleen kerran toiselle puolelle "viivaa".
