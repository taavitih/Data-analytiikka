Tästä pystyy näkemään, että jokaisen tilauksen kohdalle on täytetty myös kaikki tässä listatut tiedot. Luonnolisesti ihan kaikkialla ei ole osavaltioita (State) tai välttämättä postinumeroita saatika alueita (Territory).
Näin datan eri rivit ovat keskenään vertailukelpoisia.

```python
df["Month"]=df["Month number"].replace({1:"Jan", 2:"Feb", 3:"Mar", 4:"Apr", 5:"May", 6:"Jun", 7:"Jul", 8:"Aug", 9:"Sep", 10:"Oct", 11:"Nov", 12:"Dec"})
```

```python
df.columns = ['Order number', 'Quantity ordered', 'A price', 'Order line number',
       'Sales', 'Order date', 'Status', 'Quarter', 'Month number',
       'Year', 'Product line', 'List price', 'Product code', 'Customer name',
       'Phone', 'Address line 1', 'Address line 2', 'City', 'State',
       'Postal code', 'Country', 'Territory', 'Contact last name',
       'Contact first name', 'Deal size', "Month"]
df.head()
```

```python
df = df [['Order number', 'Quantity ordered', 'A price', 'Order line number',
       'Sales', 'Order date', 'Status', 'Quarter', 'Month number', "Month",
       'Year', 'Product line', 'List price', 'Product code', 'Customer name',
       'Phone', 'Address line 1', 'Address line 2', 'City', 'State',
       'Postal code', 'Country', 'Territory', 'Contact last name',
       'Contact first name', 'Deal size']]
df.head()
```

Kuukausia hahmottaa nyt mahdollisesti paremmin, kun numeron vieressä on myös kuukauden lyhenne.

```python
df[["Country", "City", "Sales", "Deal size"]].sort_values(by="Sales", ascending=False).head(20)
```

Suurimmat tilaukset menivät Yhdysvaltoihin.

```python
df[["Country", "City", "Sales", "Month", "Quarter"]].sort_values(by="Sales", ascending=False).head(20)
```
