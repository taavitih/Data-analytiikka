Kunnat joissa oli joko muuttotappiota vai muuttovoittoa.

```python
df[["Kunta", "Vuosikate, euroa/asukas, 2017", "Lainakanta, euroa/asukas, 2017"]].sort_values(by="Vuosikate, euroa/asukas, 2017", ascending=False).head(10)
```

Tarkastelua missä kunnissa oli paras vuosikate per asukas, mutta myös miten velkaa oli per asukas.

```python
df[["Kunta", "Yli 64-v osuus väestöstä, %, 2017", "Lainakanta, euroa/asukas, 2017", "Konsernin lainakanta, euroa/asukas, 2017"]].sort_values(by="Yli 64-v osuus väestöstä, %, 2017", ascending=False).head(10)
```

Kunnat joissa väestö on eniten ikääntynyt ja kuinka paljon lainaa per asukas. Mitä vanhempaa väestö ja mitä enemmän lainaa, niin lähtökohtaisesti sitä vaikeampi tulevaisuus, koska verojen maksajia vähemmän.
Toisin sanoen tulevaisuudessa taloudellisissa vaikeuksissa tulevat olemaan mahdollisesti Kuhmoinen, Luhanka, Vesanto, Kaskinen, Kustavi ja Padasjoki. Tämän datan avulla ei voi kuitenkaan tehdä kovinkaan luetattavia arvioita, korkeintaan suuntaa-antavia.

```python
df[["Kunta", "15-64 -v osuus väestöstä, %, 2017", "Alueella olevien työpaikkojen lkm, 2016", "Väkiluku, 2017"]].sort_values(by="15-64 -v osuus väestöstä, %, 2017", ascending=False).head(10)
```

Kunnat joissa eniten työssäkäyvää väestöä (15-64 -vuotiaita) näyttää myös olevan työpaikkoja noin puolelle väkiluvusta. Järvenpäässä, Oulussa ja Rovaniemellä kuitenkin työpaikkoja on suhteessa vähemmän, vaikka työväkeä olisi tarjolla.
